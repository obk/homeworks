```python
# Probability is a numerical description of how likely an event is to occur.
# Probability is a way of quantifying the uncertainty associated with events chosen from a some universe of events.
# Rolling a die : The universe consists of all possible outcomes. And any subset of these outcomes is an event; 
# for example, “the die rolls a one” or “the die rolls an even number.”
# P(E) to mean “the probability of the event E.”

# Tossing two coins
# {TT, TH, HT, HH}
# Total number of events(universe) 2^2
```

```python
# we say that two events E and F are dependent if knowing something about whether E happens gives us information about 
# whether F happens (and vice versa). Otherwise they are independent.
```

```python
# When two events E and F are independent: the probability is P (E,F) = P(E)*P(F)
# The probability of E “conditional on F”   P (E|F) = P (E,F) /P( F)
```

```python
from collections import Counter
import math, random
import matplotlib.pyplot as plt
```

```python
def random_kid():
    return random.choice(["boy", "girl"])
random_kid()
```

```python
# CONDITIONAL PROBABILITY

both_girls = 0
older_girl = 0
either_girl = 0

random.seed(42)
for i in range(1000):
    younger = random_kid()
    older = random_kid()
    if older == "girl":
        older_girl += 1
    if older == "girl" and younger == "girl":
            both_girls += 1
    if older == "girl" or younger == "girl":
            either_girl += 1
    if(i%100==0 and i>0):
        print("P(both | older):",i, both_girls,older_girl, both_girls / older_girl)      # 0.514 ~ 1/2
        print("P(both | either): ",i, both_girls,either_girl, both_girls / either_girl)   # 0.342 ~ 1/3
```

```python
# Imagine a certain disease that affects 1 in every 10,000 people. And imagine
# that there is a test for this disease that gives the correct result (“diseased” if you have
# the disease, “nondiseased” if you don’t) 99% of the time.
# Let’s use T for the event “your test is positive” and D for the event “you have the disease.” 
# Then Bayes’s Theorem says that the probability that you have the disease, conditional on testing positive, is

# P (D|T) = P (D,T) /P(T)  ==> P (D,T) = P (D|T) P(T)
# P (T|D) = P (T,D) /P(D)  ==> P (T,D) = P (T|D) P(D) 
# P (D,T) = P (T,D)  ==> P (D|T) P(T) = P (T|D) P(D)
# P (D|T) = P (T|D) P (D) / P(T)

# we know that P (T| D) , the probability that someone with the disease tests positive, is 0.99.
# P(D) , the probability that any given person has the disease, is 1/10,000 = 0.0001.
# P (T|not D) , the probability that someone without the disease tests positive, is 0.01. 

# P(T) = 0.010098
# P (D|T) = P (T|D) P (D) / P(T) = 0.99*0.0001/0.010098=0.0098039..= 0.98%

# people who test positive actually have the disease P (D| T) ~= 0.0098 less than 1%   ()
```

```python
# https://www.itl.nist.gov/div898/handbook/eda/section3/eda3662.htm
#probablity density function
def uniform_pdf(x):
    return 1 if x >= 0 and x < 1 else 0
print(uniform_pdf(0.4))

```

```python
def plot_uniform_pdf():
    xs = [x / 10.0 for x in range(-0, 10)]
    plt.plot(xs,[uniform_pdf(x)   for x in xs],'-',label='standard uniform distribution function')
    plt.legend()
plot_uniform_pdf()
```

```python
# cumulative distribution function
# the cumulative distribution function (cdf), which gives the probability that a random variable is less than or equal to a certain value.
def uniform_cdf(x):
    "returns the probability that a uniform random variable is less than x"
    if x < 0:   return 0    # uniform random is never less than 0
    elif x < 1: return x    # e.g. P(X < 0.4) = 0.4
    else:       return 1    # uniform random is always less than 1

print(uniform_cdf(0.4))
```

```python
def plot_uniform_cdf():
    xs = [x / 10.0 for x in range(-0, 10)]
    plt.plot(xs,[uniform_cdf(x)   for x in xs],'-',label='cumulative distribution function')
    plt.legend()
plot_uniform_cdf()
```

```python
# The normal distribution is the king of distributions. It is the classic bell curve–shaped
# distribution and is completely determined by two parameters: its mean μ (mu) and its standard deviation σ (sigma).
# The mean indicates where the bell is centered, and the standard deviation how “wide” it is.
# When μ = 0 and σ = 1, it’s called the standard normal distribution.
def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))

def plot_normal_pdfs():
    xs = [x / 10.0 for x in range(-50, 50)]
    plt.plot(xs,[normal_pdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
    plt.plot(xs,[normal_pdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
    plt.plot(xs,[normal_pdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
    plt.plot(xs,[normal_pdf(x,mu=-1)   for x in xs],'-.',label='mu=-1,sigma=1')
    plt.legend()
    plt.show()
plot_normal_pdfs()
```

```python
def normal_cdf(x, mu=0,sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2

def plot_normal_cdfs():
    xs = [x / 10.0 for x in range(-50, 50)]
    plt.plot(xs,[normal_cdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
    plt.plot(xs,[normal_cdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
    plt.plot(xs,[normal_cdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
    plt.plot(xs,[normal_cdf(x,mu=-1) for x in xs],'-.',label='mu=-1,sigma=1')
    plt.legend(loc=4) # bottom right
    plt.show()
plot_normal_cdfs()
```
