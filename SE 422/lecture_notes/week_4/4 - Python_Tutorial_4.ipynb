```python
#Write a class named Rectangle with a constructor that takes length and width variables and a function to calculate the area of a rectangle. Next, drive a Square class (inheritance) from the Rectangle class. The Square class must have the function to return a boolean value to tell whether it is a square. Also, the Rectangle and Square classes must have a "name" class variables.

class Rectangle():
    name='Rectangle'
    def __init__(self, l, w):
        self.length = l
        self.width  = w

    def __repr__(self):
        """this is the string representation of the object
            if you type it at the Python prompt or pass it to str()"""
        return "Rectangle"

    def area(self):
        return self.length*self.width
class Square(Rectangle):
    name='Square'
    def __init__(self,l,w):
        super().__init__(l,w)
    def __repr__(self):
        """this is the string representation of the object
            if you type it at the Python prompt or pass it to str()"""
        return "Square"

    def isSquare(self):
        if self.length==self.width:
            return True

newRectangle = Rectangle(12, 10)
print(newRectangle, newRectangle.name,newRectangle.area())
newSquare=Square(10,10)
#print(newRectangle,  newSquare)
print(newSquare,newSquare.name,newSquare.isSquare(),newSquare.area())
help(Square.__repr__)
```

```python
class Animal():
    name='animal'
    def __init__(self, nLegs=4,sound='some sounds', tail=True):
        self.nLegs=nLegs
        self.sound=sound
        self.tail=tail
    def run(self):
        return "animal runs fast"
class Cat(Animal):
    def __init__(self, nLegs,sound, tail):
        super().__init__(nLegs,sound, tail)
    def eats():
        return "Eats meat"

class Kangaroo(Animal):
    def __init__(self, nLegs=2,sound='', tail=True):
        super().__init__(nLegs,sound, tail)
    def eats():
        return "Eats vegetable"
    def kicks():
        return "Kangaroo kicks"
an=Animal()
cat=Cat(4,'miyav',True)
kan=Kangaroo(sound='kaaaaa',tail=True)
print(an.name, cat.name, kan.name)
```

```python
#Write a Python program that matches a string that has an H followed by one or more l's.


import re
def text_match(text):
        patterns = '^H.*l+.*'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("Helloooooo"))
print(text_match("Hi"))
```

## Try Except

The try block lets you test a block of code for errors. <br/>
The except block lets you handle the error. <br/>
The finally block lets you execute code, regardless of the result of the try- and except blocks. 

```python
try:
  print(xy)
except:
  print("An exception occurred") 
```

```python
try:
  print(z)
except NameError:
  print("Variable z is not defined")
except:
  print("Something else went wrong") 
```

```python
try:
  print("Hello")
except:
  print("Something went wrong")
else:
  print("Nothing went wrong") 
```

```python

try:
  print(ty)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished") 
```

## Visualizing data

```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='#137a8c', marker='o', linestyle='solid', linewidth=5, markersize=20)

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# bars are by default width 0.8, so we'll add 0.1 to the left coordinates
# so that each bar is centered
xs = [i for i, _ in enumerate(movies)]
print(xs)
# plot bars with left x-coordinates [xs], heights [num_oscars]
plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards")
plt.xlabel("X test")
plt.title("My Favorite Movies")
# label x-axis with movie names at bar centers
print(([i for i, _ in enumerate(movies)], movies))
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

decile = lambda grade: grade // 10*10
print(list(map(decile,grades)))

histogram = Counter(decile(grade) for grade in grades)
print(histogram)
plt.bar([x - 4 for x in histogram.keys()], # shift each bar to the left by 4
histogram.values(), # give each bar its correct height
8) # give each bar a width of 8
plt.axis([-10, 105, 0, 5]) # x-axis from -5 to 105,
# y-axis from 0 to 5
plt.xticks([50 * i for i in range(3)]) # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
# we can make multiple calls to plt.plot
# to show multiple series on the same chart
plt.plot(xs, variance, 'g-', label='variance') # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # red dot-dashed line
plt.plot(xs, total_error, 'b:', label='total error') # blue dotted line
# because we've assigned labels to each series
# we can get a legend for free
# loc=9 means "top center"
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)
plt.title("Daily Minutes vs. Number of Friends")
# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,xy=(friend_count, minute_count), xytext=(5, 0),textcoords='offset points')
plt.show()
```
